```python
import matplotlib.pyplot as plt
```

```python
a=[4,6,8]
b=[8,9,12]
plt.plot(a,b)
```

```python
plt.plot(b,a)
```

```python
plt.bar(a,b)
```

```python
# scatter plot
wt=[50,55,56,67.5,68]
ht=[123,177,144,168,156]
plt.scatter(wt,ht)
plt.xlabel("Weight")
plt.ylabel("Height")
plt.title("BMI")
```

```python
data=['india','uk','us','norway']
area=[13.5,34,44,12]
plt.pie(area,labels=data)
```

```python
e=[0.3,0,0,0]
plt.pie(area,explode=e,labels=data)
```

```python
# 3d
pic=plt.axes(projection='3d')
pic.plot3D(area,e)
```

```python
# relplot
sns.relplot(data=dftips,x='total_bill',y='tip',col='time',hue='smoker',style='smoker',size='size')
```
